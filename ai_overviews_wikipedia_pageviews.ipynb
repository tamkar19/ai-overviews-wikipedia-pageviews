# Did AI Overviews reduce Wikipedia traffic?

## Background

Google AI Overviews have changed how users interact with search results. Instead of clicking through to external websites, users may receive an answer directly on the search results page.

Wikipedia is one of the world's largest open knowledge platforms. Its articles are widely used as references in search engines, AI systems, and other information services. At the same time, Wikimedia provides open access to pageview statistics, making it possible to study user behaviour using publicly available data.

This combination makes Wikipedia an excellent case for investigating whether AI-generated answers changed how people access information.

## **Research question**

**Did the launch of Google AI Overviews reduce traffic to selected Wikipedia articles?**

## Why this dataset?

I intentionally chose Wikipedia because the data are publicly available, reproducible, and globally relevant. Anyone can rerun this analysis using the same open datasets, making the research transparent and easy to verify.


## Treatment group selection

AI Overviews are not shown for every search query. Therefore, the treatment group should consist of topics where Google frequently displays AI-generated answers and where users often seek factual information.

Instead of selecting articles randomly, I manually verified whether Google AI Overviews appeared for each candidate query.

The treatment group consists of medical topics that satisfy all of the following criteria:

- Google displays an AI Overview.
- The query is informational rather than navigational.
- A corresponding English Wikipedia article exists.
- The topic is commonly searched and suitable for public traffic analysis.

The treatment group definition is stored in a separate Excel file to make the analysis reproducible and easy to modify.

## Data collection

Wikipedia provides a public Pageviews API that allows anyone to retrieve daily traffic statistics for individual articles.

For each selected article, daily pageviews were collected from May 14, 2023 to November 14, 2024.

The intervention date was set to May 14, 2024, corresponding to Google's broader rollout of AI Overviews.

```python
START_DATE = "2023-05-14"
END_DATE = "2024-11-14"

INTERVENTION_DATE = "2024-05-14"
```

```python
ARTICLES = treatment["Wikipedia article"].tolist()

ARTICLES
```

```python
import requests
import pandas as pd

def get_pageviews(article, start_date, end_date, project="en.wikipedia.org"):
    article_encoded = article.replace(" ", "_")
    start = start_date.replace("-", "")
    end = end_date.replace("-", "")

    url = (
        f"https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/"
        f"{project}/all-access/user/{article_encoded}/daily/{start}/{end}"
    )

    headers = {
        "User-Agent": "Wikipedia AI Overviews Research by Tamara Karavaya (tamkor19@gmail.com)"
    }

    response = requests.get(url, headers=headers)

    if response.status_code != 200:
        print(f"Failed: {article} | status code: {response.status_code}")
        return pd.DataFrame()

    data = response.json()["items"]

    df = pd.DataFrame(data)
    df["date"] = pd.to_datetime(df["timestamp"].str[:8])
    df["article"] = article
    df = df[["date", "article", "views"]]

    return df
```

```python
test = get_pageviews("Migraine", START_DATE, END_DATE)
test.head()
```

```python
all_pageviews = []

for article in ARTICLES:
    print(f"Fetching: {article}")
    article_df = get_pageviews(article, START_DATE, END_DATE)
    all_pageviews.append(article_df)

raw_pageviews = pd.concat(all_pageviews, ignore_index=True)

print(f"Rows collected: {len(raw_pageviews)}")
raw_pageviews.head()
```

```python
raw_pageviews.to_csv("raw_pageviews_data.csv", index=False)

print("Raw dataset saved.")
```

## Exploratory Data Analysis

Before estimating any causal effect, I first explore the raw pageview data. The goal of this section is to understand the overall traffic pattern, identify visible trends or seasonality, and check whether there is any noticeable change around the intervention date.

At this stage, I do not make causal claims. The purpose of EDA is descriptive: to understand the data before applying causal inference methods.


```python
daily_views = (
    raw_pageviews
    .groupby("date", as_index=False)["views"]
    .sum()
)

daily_views.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 6))

plt.plot(daily_views["date"], daily_views["views"])

plt.axvline(
    pd.to_datetime(INTERVENTION_DATE),
    linestyle="--",
    label="Google AI Overviews launch"
)

plt.title("Daily pageviews for selected Wikipedia articles")
plt.xlabel("Date")
plt.ylabel("Pageviews")
plt.legend()
plt.show()
```

The aggregated time series shows daily pageviews across all selected treatment articles. The vertical dashed line marks May 14, 2024, the launch date of Google AI Overviews in the United States.

The raw daily series is noisy, which is expected for web traffic data. Before interpreting any changes around the intervention date, I apply a rolling average to better observe the underlying trend.


```python
daily_views["rolling_30d"] = (
    daily_views["views"]
    .rolling(window=30, min_periods=1)
    .mean()
)
```

```python
plt.figure(figsize=(14, 6))

plt.plot(
    daily_views["date"],
    daily_views["views"],
    alpha=0.3,
    label="Daily pageviews"
)

plt.plot(
    daily_views["date"],
    daily_views["rolling_30d"],
    linewidth=2,
    label="30-day rolling average"
)

plt.axvline(
    pd.to_datetime(INTERVENTION_DATE),
    linestyle="--",
    label="Google AI Overviews launch"
)

plt.title("Daily pageviews with 30-day rolling average")
plt.xlabel("Date")
plt.ylabel("Pageviews")
plt.legend()
plt.show()
```

### Distribution of pageviews across articles

Before aggregating all articles into a single time series, I examine how pageviews are distributed across the treatment group. This helps determine whether the aggregated trend is representative of the entire group or is dominated by a small number of highly popular articles.


```python
article_summary = (
    raw_pageviews
    .groupby("article", as_index=False)
    .agg(
        total_views=("views", "sum"),
        mean_daily_views=("views", "mean")
    )
    .sort_values("total_views", ascending=False)
)

article_summary
```

```python
plt.figure(figsize=(10, 6))

plt.barh(
    article_summary["article"],
    article_summary["total_views"]
)

plt.xlabel("Total pageviews")
plt.title("Total pageviews by article")

plt.gca().invert_yaxis()

plt.tight_layout()
plt.show()
```

## Interpretation

Pageviews are distributed across all selected articles rather than being concentrated in a single topic. Although articles such as Diabetes and Hypertension receive more traffic than others, no individual article dominates the aggregated dataset. This suggests that the overall time series reflects the combined behavior of the treatment group rather than being driven by a single high-traffic page.

## Limitations

While the selected articles are plausible candidates for appearing in Google AI Overviews, article-level exposure cannot be directly verified. The Wikimedia Pageviews API provides pageview counts but does not indicate whether visits originated from search results containing an AI Overview. Consequently, treatment assignment is based on reasonable assumptions rather than directly observed exposure.

In addition, exploratory analysis did not reveal a clear discontinuity around the intervention date. This does not imply the absence of an effect but suggests that article-level pageviews alone are insufficient for drawing strong causal conclusions.

## Conclusion

This project demonstrates a fully reproducible workflow for collecting and exploring Wikipedia pageview data using the Wikimedia Pageviews API.

The exploratory analysis did not reveal convincing visual evidence of an immediate change in pageviews following the launch of Google AI Overviews. At the same time, the analysis identified an important methodological limitation: treatment exposure cannot be directly observed at the article level using publicly available pageview data alone.

Rather than supporting a strong causal claim, these findings emphasize the importance of carefully evaluating identification strategies when working with observational data.
